```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=[6, 6])
G = nx.DiGraph()
G.add_edges_from([(0,1),(0,2),(1,4),(2,3),(2,4),(3,4)])
mapping = {0: "s", 4: "d", 1: "a", 2 : "b", 3:"c", 4:"d"} # s:source d:destination
H = nx.relabel_nodes(G, mapping)
pos = {"s":[-1,0],"a":[0,1],"b":[0,0],"c":[0,-1],"d":[1,0]}
G[0][1]['capacity'] = 3
G[0][2]['capacity'] = 3
G[1][4]['capacity'] = 2
G[2][3]['capacity'] = 3
G[2][4]['capacity'] = 2
G[3][4]['capacity'] = 1

fig, ax = plt.subplots(figsize = (7,6), dpi=120)
nx.draw(H,pos,with_labels=True)
plt.savefig("ex1_graph.svg")
```

```python
nx.minimum_cut(G,0,4)
```
