```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline

G = nx.DiGraph()
G.add_edges_from([(0,1),(0,2),(1,4),(2,3),(2,4),(3,4)])
mapping = {0: "s", 4: "d", 1: "a", 2 : "b", 3:"c", 4:"d"} # s:source d:destination
H = nx.relabel_nodes(G, mapping)
pos = {"s":[-1,0],"a":[0,1],"b":[0,0],"c":[0,-1],"d":[1,0]}


nx.draw(H,pos,with_labels=True)
```

```python
tmp = nx.all_simple_paths(G,source = 0, target = 4)
for el in tmp:
  print(el)
```

```python
tmp = nx.node_disjoint_paths(G, s=0, t=4)
for el in tmp:
  print(el)
```

```python
G.remove_edge(1, 4)
G.remove_edge(0,2)
mapping = {0: "s", 4: "d", 1: "a", 2 : "b", 3:"c", 4:"d"} # s:source d:destination
H = nx.relabel_nodes(G, mapping)
pos = {"s":[-1,0],"a":[0,1],"b":[0,0],"c":[0,-1],"d":[1,0]}


nx.draw(H,pos,with_labels=True)
```

```python
CG = nx.algorithms.components.condensation(G)
print(dict(CG.nodes))
```
